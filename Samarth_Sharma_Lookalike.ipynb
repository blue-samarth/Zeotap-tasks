```python
import numpy as np
import pandas as pd
```

```python
df_data : pd.DataFrame = pd.read_csv('data/df_all.csv')
df_data
```

```python
categorical_cols : list[str] = ["Region", "FavCategory",
                                "MostActiveTransactionDay", "PeakTransactionTime",
                                "TransactionMonth", "TransactionDay",
                                "Transaction_TimePeriod", "Category",
                                "FavProduct"]
df_encoded = pd.get_dummies(df_data, columns=categorical_cols, drop_first=True)
df_encoded
```

```python
from sklearn.preprocessing import StandardScaler

numeric_cols : list[str] = ["AvgTransactionValue", "DaysSinceLastTransaction", "UniqueProductsPurchased", "TotalValue", "Quantity", "Price"]
scaler = StandardScaler()
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])
df_encoded
```

```python
df_encoded.columns
```

```python
final_features : pd.DataFrame = df_encoded.drop([
    "TransactionID", "CustomerID", "ProductID", "TransactionDate",
    "ProductName", "CustomerName", "SignupDate", "TransactionTime"
], axis=1)
final_features

final_features.to_csv("final_features.csv", index=False)
```

```python
from sklearn.neighbors import NearestNeighbors

model : NearestNeighbors = NearestNeighbors(n_neighbors=10, algorithm='auto', metric='cosine')
model.fit(final_features)
```

```python
def get_similar(target_customer_id: str) -> list[tuple[str, float]]:
  try:
    target_indx: np.int64 = df_encoded[df_encoded["CustomerID"] == target_customer_id].index[0]
    distance: np.ndarray
    indices: np.ndarray
    distance, indices = model.kneighbors(final_features.iloc[target_indx].values.reshape(1, -1))
    similarities: list = 1 - distance.flatten()

    neighbour_ids: list = df_encoded.iloc[indices.flatten(), :]["CustomerID"].tolist()

        # Filter out the target customer and create list of tuples
    filtered_results = [
        (cust_id, sim)
        for cust_id, sim in zip(neighbour_ids, similarities)
        if cust_id != target_customer_id
    ]

    return filtered_results
  except IndexError:
    return []
```

```python
import pandas as pd
from typing import List, Tuple, Optional

def print_similar_customers(customer_id: str, df: pd.DataFrame, row_index: int, top_n: int = 3) -> None:
    """
    Prints and stores the most similar customers for a given customer ID, excluding the customer itself.

    Args:
        customer_id (str): The target customer ID (e.g., 'C0001')
        df (pd.DataFrame): DataFrame to store the results
        row_index (int): Index in the DataFrame where results should be stored
        top_n (int): Number of similar customers to display (default: 3)
    """
    results = list(get_similar(customer_id))

    if not results:
        print(f"No results found for customer {customer_id}")
        return


    filtered_results = [
        (similar_id, score)
        for similar_id, score in results
        if similar_id != customer_id
    ][:top_n]


    df.at[row_index, 'CustomerID'] = customer_id

    for i, (similar_id, score) in enumerate(filtered_results, 1):
        df.at[row_index, f'Most_Similar_{i}'] = similar_id
        df.at[row_index, f'Similarity_Score_{i}'] = score

def create_similarity_analysis(start_id: int = 1, end_id: int = 20, top_n: int = 3) -> pd.DataFrame:
    """
    Creates a DataFrame with similar customers for a range of customer IDs.

    Args:
        start_id (int): Starting customer number (default: 1)
        end_id (int): Ending customer number (default: 20)
        top_n (int): Number of similar customers to find for each customer (default: 3)

    Returns:
        pd.DataFrame: DataFrame containing similarity analysis results
    """

    columns = ['CustomerID']
    for i in range(1, top_n + 1):
        columns.extend([f'Most_Similar_{i}', f'Similarity_Score_{i}'])


    df = pd.DataFrame(columns=columns)


    for i in range(start_id, end_id + 1):
        customer_id = f'C{str(i).zfill(4)}'
        print_similar_customers(customer_id, df, i - start_id, top_n)

    return df
```

```python
similarity_df = create_similarity_analysis(1, 20, 3)
similarity_df
```

```python
def save_to_csv(df: pd.DataFrame, filename: str) -> bool:
    """
    Saves a DataFrame to a CSV file.

    Args:
        df (pd.DataFrame): DataFrame to be saved
        filename (str): Name of the CSV file to save to
    """
    try:
      df.to_csv(filename, index=False)
      return True
    except:
      print("Error saving to CSV")
      return False
```

```python
save_to_csv(similarity_df, "data/Lookalike.csv")
```
